## **Problem Statment**
How efficiently do NBA teams convert their payroll into winning outcomes, and which specific player combinations offer the best return on investment in real-game contexts?

#### **Detailed Version**
In the NBA, team success is often associated with high payrolls, but financial investment doesn't always guarantee winning results. With salary caps and luxury taxes shaping team strategies, it becomes crucial to understand how efficiently franchises convert their spending into on-court performance. This project explores the relationship between team payrolls and regular season success, aiming to uncover whether financial outlay truly correlates with winning or if some teams consistently outperform expectations through smarter spending and roster management.

## **Research Questions**

- How expensive is a regualr season win, and how has that cost evolved over time? (cost per win = payroll/wins)
- Are high payroll teams always the most succesful in terms of regular season wins (payroll rank and win rank)
- Which teams achieve the highest ROI per dolalr spent on player performance? (roi = plus_minus total/ team payroll)
- Do some teams consistently outperfrom their payroll
- Does spending on different positions pay off equally?
- Are younger players better bargains than veterans
- How does average ROI per position evolve across NBA eras?

```python
# display merged DataFrame
merge2
```

```python
# Compute total salary and net plus_minus per team each season
team_agg = merge2.groupby(["start_year","team_name"]).agg({
    "salary": "sum",
    "plus_minus": "sum"
}).reset_index()  # DataFrame with total salary and plus_minus by team and season
team_agg  # display aggregated team metrics
```

```python
# Grouping by season, player, and salary then compute total plus_minus and sort by salary descending
merge3 = merge2.groupby(["start_year","player_name","salary"])[["plus_minus"]].sum().reset_index().sort_values(["start_year","salary"], ascending=[True,False])  # DataFrame with net plus_minus per player per season sorted by salary
merge3  # displaying the aggregated player metrics
```

```python
p_info =  player_info.copy()

# Step 1: Create a list of years for each row
p_info['season_years'] = p_info.apply(
    lambda row: list(range(row['season_start_year'], row['season_end_year'] + 1)),
    axis=1
)

# Step 2: Explode the DataFrame so each season gets its own row
expanded_df = p_info.explode('season_years', ignore_index=True)

# Step 3: Rename the exploded column
expanded_df = expanded_df.rename(columns={'season_years': 'start_year'})

# Step 4: Drop original start/end columns (optional)
expanded_df = expanded_df.drop(columns=['season_start_year', 'season_end_year'])

# Final result: one row per player per season
display(expanded_df)
```

```python
# Merge aggregated player metrics with player positions by season and name
player_pos_data = merge3.merge(
    expanded_df[["start_year","player_name","position"]],
    on=["start_year","player_name"]
)  # DataFrame combining plus_minus metrics with positional info
player_pos_data  # display merged DataFrame
```

#### **Adding More Features to the Merged Dataset**

```python
# Incorporate total home wins into team metrics
team_agg["home_win"] = total_wins_per_team["home_win"]  # add home win counts per team

# Compute cost per win in millions
team_agg["cost_per_win(M)"] = team_agg["salary"] / (team_agg["home_win"] * 1e6)  # salary divided by wins (Millions)

# Rank teams by salary and wins within each season
team_agg["salary_rank"] = team_agg.groupby("start_year")["salary"].rank(method="dense", ascending=False).astype(int)  # highest salary gets rank 1
team_agg["win_rank"] = team_agg.groupby("start_year")["home_win"].rank(method="dense", ascending=False).astype(int)  # most wins gets rank 1

# Calculate return on investment (ROI) per team
team_agg['roi'] = team_agg['plus_minus'] / team_agg['salary'] * 1e6  # net plus_minus per million dollars spent
roi_min = team_agg['roi'].min()  # minimum ROI value
roi_max = team_agg['roi'].max()  # maximum ROI value
team_agg['roi_normalized'] = (team_agg['roi'] - roi_min) / (roi_max - roi_min)  # scale ROI between 0 and 1
```

```python
# displaying first few rows of team_agg
team_agg.head()  
```

```python
# Compute ROI for each player-season (net plus_minus per million dollars spent)
player_pos_data["roi"] = player_pos_data['plus_minus'] / player_pos_data['salary'] * 1e6

# Aggregate ROI by season and position
player_pos_data_filtered = player_pos_data.groupby(
    ["start_year", "position"]
)[["roi"]].sum().reset_index()

# Keep only positions with positive ROI
player_pos_data_filtered = player_pos_data_filtered[
    player_pos_data_filtered["roi"] > 0
]

# Display the position-level ROI results
player_pos_data_filtered
```

```python
# Sorting by player name and season to ensure chronological order
player_pos_data = player_pos_data.sort_values(by=['player_name', 'start_year'])  # order records for experience calculation

# Compute the number of seasons each player has played up to each season
player_pos_data['experience'] = (
    player_pos_data
    .groupby('player_name')
    .cumcount() + 1
)  # experience as season count

player_pos_data  # display DataFrame with experience feature
```

#### **ROI Share by Position (Donut Charts per Era)**

```python
# define function to map a start year to an era label
def assign_era(year):
    # years 1996 through 2009 go to Early Era
    if 1996 <= year <= 2009:
        return 'Early Era'
    # years 2010 through 2016 go to Last Decade
    elif 2010 <= year <= 2016:
        return 'Last Decade'
    # years 2017 through 2019 go to Current Era
    elif 2017 <= year <= 2019:
        return 'Current Era'
    # all other years are excluded
    return None

# copy filtered data so original is not modified
era_data = player_pos_data_filtered.copy()
# apply era assignment to each row based on start_year column
era_data['era'] = era_data['start_year'].apply(assign_era)
# remove any rows where assign_era returned None
era_data = era_data[era_data['era'].notna()]

# group by era and position, summing the roi values
era_grouped = era_data.groupby(['era', 'position'], as_index=False)['roi'].sum()

# list of eras in desired plotting order
eras = ['Early Era', 'Last Decade', 'Current Era']
# create one subplot per era, arranged in a row
fig, axes = plt.subplots(1, len(eras), figsize=(12, 4))

for ax, era in zip(axes, eras):
    # select data for this specific era
    data = era_grouped[era_grouped['era'] == era]
    # choose one color per position from seaborn's Set2 palette
    colors = sns.color_palette('Set2')[:len(data)]
    
    # draw a donut chart: pie chart with inner radius
    ax.pie(
        data['roi'],             # wedge sizes
        labels=None,             # no labels on wedges
        autopct='%1.1f%%',       # show percentage inside each wedge
        startangle=90,           # start first wedge at 12 o'clock
        wedgeprops={'width': 0.4},  # width makes it a donut
        colors=colors,           # color list for wedges
        pctdistance=0.75         # position of percentage labels
    )
    # set title to the era name
    ax.set_title(era)
    # ensure the pie is drawn as a circle
    ax.axis('equal')

# create legend handles mapping each position to its color
legend_handles = [
    Patch(facecolor=col, label=pos)
    for pos, col in zip(era_grouped['position'].unique(), sns.color_palette('Set2'))
]
# add a shared legend to the figure on the right side
fig.legend(handles=legend_handles, loc='center right', title='Position')
# adjust layout to make room for legend
plt.tight_layout(rect=[0, 0, 0.85, 0.95])
# add a main title for the entire figure
plt.suptitle('ROI Share by Position', fontsize=16)
# display the plot
plt.show()
```
